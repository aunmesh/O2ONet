```python
f_path = '/workspace/work/misc/O2ONet/sota_experiments/zero-shot/labels.pkl'

import pickle as pkl

ooi_data = pkl.load(open(f_path, 'rb'))

import json

coin_path = open('/workspace/work/misc/O2ONet/sota_experiments/zero-shot/coin.json','rb')
coin_data = json.load(coin_path)
```

```python
def add_subaction(ooi_video_data, coin_data):
    
    fps = 25.0
    yt_id = ooi_video_data['metadata']['yt_id']
    frame_no = ooi_video_data['metadata']['frame no.']
    time_loc = float(frame_no)/25.0
    
    subactions = coin_data['database'][yt_id]['annotation']
    
    for s in subactions:
        lower_limit, upper_limit = s['segment']
        if time_loc >= lower_limit and time_loc <= upper_limit:
            return s['label']
    
    return -1

def master_function(ooi_data, coin_data):
    
    for i, o in enumerate(ooi_data):
        temp_subaction = add_subaction(o, coin_data)
        ooi_data[i]['metadata']['subaction'] = temp_subaction
        if temp_subaction == -1:
            print("Issue in ", i)
    
    return ooi_data

ooi_data = master_function(ooi_data, coin_data)
```

```python
yt_ids = [ ooi_data[i]['metadata']['yt_id'] for i in range(len(ooi_data)) ]

activities = [ ooi_data[i]['metadata']['activity name'] for i in range(len(ooi_data)) ]

from collections import Counter

print(Counter(activities).keys()) # equals to list(set(words))
print( len(Counter(activities).keys()) ) # equals to list(set(words))
print(Counter(activities).values()) # counts the elements' frequency
```
